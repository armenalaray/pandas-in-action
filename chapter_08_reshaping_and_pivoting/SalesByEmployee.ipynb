```python
import pandas as pd
```

A wide data set is ideal for seeing the aggregate picture

```python
sales = pd.read_csv("sales_by_employee.csv", parse_dates=["Date"])
```

```python
sales.info()
```

```python
sales.pivot_table(index="Date", values=["Revenue"], aggfunc="sum", columns="Name", margins=True)
```

```python
sales.pivot_table(index="Date", values=["Revenue"], aggfunc="max", columns="Name", margins=True, fill_value=0, margins_name="Total")
```

```python
sales.pivot_table(index="Date", values="Revenue", aggfunc=["sum","count"], columns="Name", fill_value=0)
```

```python
sales.pivot_table(index="Date", values=["Revenue","Expenses"], columns="Name", aggfunc=["sum","count"], fill_value=0)
```

```python
sales.pivot_table(index="Date", values=["Revenue","Expenses"], columns="Name", aggfunc=["sum","count","max","min"], fill_value=0)
```

```python
sales.pivot_table(
    index="Date", 
    values=["Revenue","Expenses"], 
    columns="Name", 
    aggfunc={"Revenue":"min","Expenses":"max"}, 
    fill_value=0
)
```

```python
sales.pivot_table(index=["Date","Name"],values=["Revenue"], aggfunc="sum")
```

```python
by_name_and_date = sales.pivot_table(
    index="Name", 
    columns="Date", 
    values="Revenue",
    aggfunc="sum"
)

by_name_and_date.stack()
```

```python
sales_by_customer = sales.pivot_table(
    index=["Customer","Name"],
    values="Revenue",
    aggfunc="sum"
)
sales_by_customer.unstack()
```
