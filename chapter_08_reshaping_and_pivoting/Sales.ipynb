```python
import pandas as pd
```

```python
video_game_sales = pd.read_csv("video_game_sales.csv")
```

aggregated!

```python
video_game_sales
```

```python
video_game_sales_by_region = video_game_sales.melt(id_vars="Name", value_vars=["NA","EU","JP","Other"], var_name="Region", value_name="Sales")
```

```python
video_game_sales_by_region
```

```python
video_game_sales_by_region.pivot_table(index="Name", values="Sales", aggfunc="sum")
```

```python
cars = pd.read_csv("used_cars.csv")
```

```python
cars.info()
```

```python
cars.pivot_table(
    index="Fuel",
    values="Price",
    aggfunc="sum"
)
```

```python
cars.pivot_table(
    index="Manufacturer", 
    columns="Transmission", 
    values="Price", 
    aggfunc="count", 
    fill_value=0, 
    margins=True, 
    margins_name="Total"
)
```

```python
cars.info()
```

```python
report = cars.pivot_table(
    index=["Year","Fuel"], 
    columns=["Transmission"], 
    values="Price", 
    aggfunc="mean"
)
```

```python
report.stack()
```
