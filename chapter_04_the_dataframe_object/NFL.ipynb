```python
nfl = pd.read_csv("nfl.csv", parse_dates=["Birthday"])
nfl["Birthday"]



nfl = nfl.set_index(keys="Name")
```

who are the five highest paid players?

```python
nfl["Salary"].sort_values(ascending=False).head()
```

```python
nfl.sort_values(by=["Team", "Salary"], ascending=[True, False]).loc[:,["Team", "Salary", "Position", "Birthday"]]
```
