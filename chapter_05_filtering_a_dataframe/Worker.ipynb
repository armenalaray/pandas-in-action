```python
import pandas as pd
```

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"])
```

```python
employees.info()
```

```python
employees["Mgmt"] = employees["Mgmt"].astype(bool)
```

```python
employees.info()
```

```python
employees["Salary"] = employees["Salary"].fillna(0).astype(int)
```

```python
employees.info()
```

```python
employees.nunique()
```

```python
employees["Gender"] = employees["Gender"].astype("category")
```

```python
employees.info()
```

```python
employees["Team"] = employees["Team"].astype("category")
```

```python
employees.info()
```

```python
employees["First Name"] == "Maria"
```

```python
employees[employees["First Name"] == "Maria"]
```

```python
employees[employees["Team"] != "Finance"]
```

```python
employees[employees["Mgmt"]]
```

```python
employees[employees["Salary"] > 100000]
```

```python
is_female = employees["Gender"] == 'Female'
in_biz_dev = employees["Team"] == "Business Dev"

a = (is_female & in_biz_dev & employees["Mgmt"])

employees[a]
```

```python
earning = employees["Salary"] < 40000
started = employees["Start Date"] > "2015-01-01"

employees[earning | started]
```

```python
a = ~(employees["Salary"] >= 100000)

employees[a]
```

```python
employees[employees["Team"].isin(["Sales","Legal","Marketing"])]
```

```python
employees[employees["Salary"].between(80000,90000)]
```

```python
employees[employees["Start Date"].between(left="1980-01-01", right="1990-01-01")]
```

```python
employees[employees["First Name"].between("R","S")]
```
